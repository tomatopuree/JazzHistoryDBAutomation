```python
from tkinter import * 
```

```python
def input_field_gen(numpages):

    Sublist = []
    
    for i in range(0,numpages):
        
        numentries = 8
        
        L2 = Label(root, text="Full Size Image Name: ").grid(row=int((i*numentries*0.5)+1), column=0, sticky=W, padx=4)
        
        Sublist.append(Entry(root))
        
        L3 = Label(root, text="Thumbnail Image Name: ").grid(row=int((i*numentries*0.5)+1), column=2, sticky=W, padx=4)
        
        Sublist.append(Entry(root))
        
        L35 = Label(root, text="Image Tag: ").grid(row=int((i*numentries*0.5)+2), column=0, sticky=W, padx=4)
        
        Sublist.append(Entry(root))
        
        L4 = Label(root, text="Attribution: ").grid(row=int((i*numentries*0.5)+2), column=2, sticky=W, padx=4)
        
        Sublist.append(Entry(root))
        
        L5 = Label(root, text="Original Image Link: ").grid(row=int((i*numentries*0.5)+3), column=0, sticky=W, padx=4)
        
        Sublist.append(Entry(root))
        
        L6 = Label(root, text="Original Image Link Source: ").grid(row=int((i*numentries*0.5)+3), column=2, sticky=W, padx=4)
        
        Sublist.append(Entry(root))
        
        L7 = Label(root, text="License Link: ").grid(row=int((i*numentries*0.5)+4), column=0, sticky=W, padx=4)
        
        Sublist.append(Entry(root))
        
        L8 = Label(root, text="License Name: ").grid(row=int((i*numentries*0.5)+4), column=2, sticky=W, padx=4)
        
        Sublist.append(Entry(root))
            
    for i in range(0,numpages):
        Sublist[(i*numentries)+0].grid(row=int((i*numentries*0.5)+1), column=1, sticky=E, pady=4)
        Sublist[(i*numentries)+1].grid(row=int((i*numentries*0.5)+1), column=3, sticky=E, pady=4)
        Sublist[(i*numentries)+2].grid(row=int((i*numentries*0.5)+2), column=1, sticky=E, pady=4)
        Sublist[(i*numentries)+3].grid(row=int((i*numentries*0.5)+2), column=3, sticky=E, pady=4)
        Sublist[(i*numentries)+4].grid(row=int((i*numentries*0.5)+3), column=1, sticky=E, pady=4)
        Sublist[(i*numentries)+5].grid(row=int((i*numentries*0.5)+3), column=3, sticky=E, pady=4)
        Sublist[(i*numentries)+6].grid(row=int((i*numentries*0.5)+4), column=1, sticky=E, pady=4)
        Sublist[(i*numentries)+7].grid(row=int((i*numentries*0.5)+4), column=3, sticky=E, pady=4)
        
    return Sublist
```

```python
def do_the_thing(numpages, list_of_things):
    
    final_str = ""
    
    beginning_html = """
    <!--Beginning of Image Slider-->
    <div id="oc-portfolio-sidebar" class="owl-carousel carousel-widget" data-items="1" data-margin="10" data-loop="true" data-nav="false" data-autoplay="5000" style="padding-top: 100px;">
    
    """
    
    final_str += beginning_html
    
    for i in range(0,numpages):
        
        numentries = 8
        
        fullsize = list_of_things[(i*numentries)+0]
        thumbsize = list_of_things[(i*numentries)+1]
        imtag = list_of_things[(i*numentries)+2]
        attrb = list_of_things[(i*numentries)+3]
        orgimlink = list_of_things[(i*numentries)+4]
        orgimlinksource = list_of_things[(i*numentries)+5]
        licensename = list_of_things[(i*numentries)+6]
        licenselink = list_of_things[(i*numentries)+7]

        single_photo_html = """
        <!-- One image in the image slider -->
        <div class="oc-item">
                <div class="iportfolio">
                    <div class="portfolio-image">
                        <!-- Replace href with full size image link-->
                        <a class="resize" href="{FSIMG}">
                            <!-- Replace src with thumbnail size image link, replace alt tag with image description -->
                            <img class="resize" src="{THMBIMG}" alt="{ALTTAG}">
                        </a>
                        <div class="portfolio-overlay">
                            <!-- Replace href with full size image link -->
                            <a href="{FSIMG}" class="center-icon" data-lightbox="image"><i class="icon-line-plus"></i></a>
                        </div>
                    </div>
                    <!-- Description of the image -->
                    <div class="portfolio-desc center nobottompadding">
                        <!-- Replace href with full size image link, replace a tag content with image description (same image and tag as above must be used) -->
                        <h3><a href="{FSIMG}" data-lightbox="image">{ALTTAG}</a></h3>
                        <!-- If there is attribution, put text (first arg below) and original image link (third arg below)-->
                        <p>{ATTRB} <a href="{LICLINK}">{LICNAME}</a>, <a href="{ORGIMLINK}">{ORGIMSRC}</a></p>
                    </div>
                </div>
            </div>
            
            
        """.format(FSIMG = fullsize, THMBIMG = thumbsize, ALTTAG = imtag, ATTRB = attrb, ORGIMLINK=orgimlink, ORGIMSRC=orgimlinksource, LICLINK=licenselink, LICNAME=licensename)
        
        final_str += single_photo_html
    

    Html_file= open("photo-slider-template.html","w")
    Html_file.write(final_str)
    Html_file.close()
    
    return 0
```

```python
### WEDNESDAY 12:30 in the lab
# ada
root = Tk()
root.title("Jazz History DB Slider Widget Tool")

Sublist = []

def get_numpages(event):
    numpages = int(numpagesE.get())
    numpagesE.delete(0, "end")
    #print("num of pages " + str(numpages))
    Sublist = input_field_gen(numpages)

    def finish(event):
        datalist = []
        for i in range(0,len(Sublist)):
            datalist.append(Sublist[i].get())
        do_the_thing(numpages, datalist)
    
    finalB = Button(root, text="Final Submit")
    finalB.grid(row=(numpages*8)+1, column=0)
    finalB.bind("<Button-1>", finish)
    

Label(root, text="Number of photos: ").grid(row=0, sticky=W, padx=4)
    
numpagesE = Entry(root)
numpagesE.grid(row=0, column=1, sticky=E, pady=4)

generatepageB = Button(root, text="Submit")
generatepageB.grid(row=0, column=2)
generatepageB.bind("<Button-1>", get_numpages)

root.mainloop()
```
